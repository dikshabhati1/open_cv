```
eye_cascade=cv2.CascadeClassifier('haarcascade_eye.xml')
```

```
face_cascade=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```
smile_cascade=cv2.CascadeClassifier('haarcascade_smile.xml')
```

```
#load image
image=files.upload()
```

**Face and eye detection**

```
# Load our image then convert it to grayscale
image = cv2.imread('diksha.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Our classifier returns the region of interest of the detected face as a tuple
# It stores the top left coordinate and the bottom right coordiantes
faces = face_cascade.detectMultiScale(gray, 2, 5)

for (fx,fy,fw,fh) in faces:
    #choose a frame of red color for the face
    cv2.rectangle(image, (fx,fy), (fx+fw,fy+fh), (0,0,255), 2)
    roi_gray = gray[fy:fy+fh, fx:fx+fw]
    roi_color = image[fy:fy+fh, fx:fx+fw]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        #choose a frame of blue color for the eyes
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(255,0,0),2)
cv2_imshow( image)     
```

**Face and smile detection**

```
# Load our image then convert it to grayscale
image = cv2.imread('diksha.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Our classifier returns the region of interest of the detected face as a tuple
# It stores the top left coordinate and the bottom right coordiantes
faces = face_cascade.detectMultiScale(gray, 2.03, 5)

for (fx,fy,fw,fh) in faces:
    #choose a frame of red color for the face
    cv2.rectangle(image, (fx,fy), (fx+fw,fy+fh), (0,0,255), 2)
    roi_gray = gray[fy:fy+fh, fx:fx+fw]
    roi_color = image[fy:fy+fh, fx:fx+fw]
    smile = smile_cascade.detectMultiScale(roi_gray)
    for (sx,sy,sw,sh) in smile:
        #choose a frame of green color for the smile
        cv2.rectangle(roi_color,(sx,sy),(sx+sw,sy+sh),(0,255,0),2)
cv2_imshow( image)   
```
